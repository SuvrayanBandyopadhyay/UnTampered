```python
import os
import re
import shutil
import tqdm
```

```python
SRC_DIR = "comofod_small/CoMoFoD_small_v2"
ORIG_DIR = "dataset/original"
TAMP_DIR = "dataset/tampered"
MASK_DIR = "dataset/mask"
```

```python
os.makedirs(ORIG_DIR, exist_ok=True)
os.makedirs(TAMP_DIR, exist_ok=True)
os.makedirs(MASK_DIR, exist_ok=True)

pattern_original = re.compile(r"^\d+_O.png$", re.IGNORECASE)
pattern_tampered = re.compile(r"^\d+_F.png$", re.IGNORECASE)
pattern_mask     = re.compile(r"^\d+_B.png$", re.IGNORECASE)

for filename in tqdm.tqdm(os.listdir(SRC_DIR)):
    src_path = os.path.join(SRC_DIR, filename)

    if not os.path.isfile(src_path):
        continue

    if pattern_original.match(filename):
        shutil.copy2(src_path, os.path.join(ORIG_DIR, filename))

    elif pattern_tampered.match(filename):
        shutil.copy2(src_path, os.path.join(TAMP_DIR, filename))

    elif pattern_mask.match(filename):
        shutil.copy2(src_path, os.path.join(MASK_DIR, filename))

print("Copying completed.")
```
